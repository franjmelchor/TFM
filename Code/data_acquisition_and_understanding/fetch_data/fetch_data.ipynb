```python
import pandas as pd
```

Datos relacionados con la beca

```python
path_data = '../../../Data/'
sep = '|'
```

```python
path_scolarship = path_data + '/Raw/Politecnica -becario por curso academico.xls'
scolarship = pd.concat(pd.read_excel(
                path_scolarship,
                header=0,
                sheet_name=None
            ), ignore_index=True)
```

```python
scolarship.head()
```

```python
columns = list(scolarship.columns)
columns.remove("expediente")
columns.insert(0, "expediente")
scolarship = scolarship[columns]
```

```python
scolarship.head()
```

```python
path_scolarship_preprocessed = path_data + '/Pre-processed/scolarship.csv'
scolarship.to_csv(
    path_scolarship_preprocessed,
    sep=sep,
    index=False)
```
