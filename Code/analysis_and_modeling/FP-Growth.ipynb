```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
import pandas as pd
```

```python
pr_personal_data = pd.read_csv('../../Data/Processed/pr_personal_data.csv', sep = '|')
pr_personal_data.head()
```

```python
pr_personal_data['fecha_curso'] = '2016-01-01'
pr_personal_data['fecha_curso'] = pd.to_datetime(pr_personal_data['fecha_curso'])
pr_personal_data['fecha_nacimiento'] = pd.to_datetime(pr_personal_data['fecha_nacimiento'])
```

```python
pr_personal_data['edad_acceso'] = pr_personal_data.apply(lambda func: func.fecha_curso.year - func.fecha_nacimiento.year,axis=1)
```

```python
pr_personal_data.head()
```

```python
final_analys_personal_data = pd.read_csv('../../Data/For_analysis_and_modeling/1st_quadrimester/pred_analys_personal_data.csv', sep = '|')
```

```python
final_analys_personal_data.drop(['1st_model','2nd_model'],axis=1,inplace=True)
```

```python
final_analys_personal_data.head()
```

```python
final_analys_personal_data['abandona'].value_counts()
```

```python
final_analys_personal_data['abandona'] = final_analys_personal_data['abandona'].astype('bool')
```

```python
final_analys_personal_data['abandona'] = final_analys_personal_data['abandona'].astype('str')
```

```python
final_analys_personal_data['abandona'].value_counts()
```

```python
dset = pd.get_dummies(data=final_analys_personal_data)
```

```python
dset.head()
```

```python
dset.drop(['expediente','cod_plan'],axis=1,inplace=True)
```

```python
pd.set_option('display.max_colwidth',-1)
```

```python
from mlxtend.frequent_patterns import fpgrowth
frequent_itemsets = fpgrowth(dset, min_support=0.15, use_colnames=True)
frequent_itemsets[0:11]
```

```python
frequent_itemsets['support'] = frequent_itemsets.apply(
    lambda func: func.support if 'abandona' in str(func.itemsets) else -1, axis=1)
frequent_itemsets_target = frequent_itemsets[frequent_itemsets['support']!=-1]
frequent_itemsets_target
```

```python
dset_apriori = apriori(dset, use_colnames = True, verbose = 1)
df_ar = association_rules(dset_apriori, metric = "confidence", min_threshold = 0.4)
df_ar
```
